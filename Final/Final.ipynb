# Hough Line Transform

To verify edges, I did the Canny Detection.

```python
#http://opencv-python-tutroals.readthedocs.org/en/latest/py_tutorials/py_imgproc/py_canny/py_canny.html

import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("/Users/intaehwang/Desktop/project2/python/MAT201A/Final/pic/final/fil33.jpg",0)
edges = cv2.Canny(img,5,20)


plt.imshow(edges,cmap = 'gray')


plt.show()
```

Based on detected edges, I run the HoughLines. minLineLength and MaxLineGap decide how many lines are drew in the picture, line function in OpenCV is straight line, thus, to make the stroke more delicate, I put random numbers. 

```python
#http://opencv-python-tutroals.readthedocs.org/en/latest/py_tutorials/py_imgproc/py_houghlines/py_houghlines.html

img = cv2.imread("/Users/intaehwang/Desktop/project2/python/MAT201A/Final/pic/final/fil33.jpg")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray,5,20,apertureSize = 3)

lines = cv2.HoughLines(edges,1,np.pi/180,200)
for rho,theta in lines[0]:
    a = np.cos(theta)*sin(theta)
    b = np.sin(theta)*cos(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 1000*(-b))
    y1 = int(y0 + 1000*(a))
    x2 = int(x0 - 1000*(-b))
    y2 = int(y0 - 1000*(a))


minLineLength = 400
maxLineGap = 5
lines = cv2.HoughLinesP(edges,1,np.pi/10,10,minLineLength,maxLineGap)

#x1 = 2000*((200*sin(x2+randint(1,19)),200*cos(x2+2))+(200*cos(x2+2),200*sin(x2+2)))
#y2 = sin(x2)

for x1,y1,x2,y2 in lines[0]:
    cv2.line(img,(x1+randint(-55,55),y1+randint(-66,66)),(x2+randint(-88,88),y2+randint(-100,100)),(255,244,255),randint(33,66))
    cv2.line(img,(x1+randint(-55,55),y1+randint(-66,66)),(x2+randint(-99,99),y2+randint(-100,100)),(116,19,115),randint(33,66))
    #cv2.line(img,(x1+randint(-55,55),y1+randint(-55,55)),(x2+randint(-33,33),y2+randint(-15,15)),(111,111,111),26)
    #cv2.line(img,(x1+randint(-22,22),y1+randint(-22,22)),(x2+randint(-22,22),y2+randint(-10,10)),(11,11,11),8)
    
#cv2.imwrite("/Users/intaehwang/Desktop/coral9c.jpg",img) 

#blur = cv2.medianBlur(img,11)


cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
