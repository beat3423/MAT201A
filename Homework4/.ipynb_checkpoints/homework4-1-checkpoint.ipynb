```python
o = imread('/Users/intaehwang/Desktop/leo1.png')
imshow(o)
colorbar()
```

```python
o = o.astype(float).sum(axis=-1)/3
o = where(o > 0.4, 0.9,-0.9)
imshow(o, cmap=cm.gray, interpolation='nearest')
```

```python
import scipy.signal
```

```python
from scipy.signal import correlate2d
cc = correlate2d(i, o)
imshow(cc)
colorbar()
gcf().set_figheight(8)
```

```python
imshow(cc, cmap=cm.gray)
colorbar()
gcf().set_figheight(8)
```

```python
imshow(where(cc > 200, 1, 0), interpolation='nearest', cmap=cm.gray)
colorbar()
gcf().set_figheight(8)
```

```python
imshow(where(cc > 1000, 1, 0), interpolation='nearest', cmap=cm.gray)
colorbar()
gcf().set_figheight(8)
```

```python
from scipy.ndimage.filters import maximum_filter
imshow(maximum_filter(cc, (10,10)))
gcf().set_figheight(8)
```

```python
subplot(121)
imshow(maximum_filter(cc, (50,50)))
subplot(122)
imshow(i, cmap=cm.gray)
gcf().set_figheight(8)
```
