```python
%pylab inline
import wave
fp = wave.open("/Users/intaehwang/Desktop/project2/python/MAT201A/Homework2/colano.wav",'r')
print(fp.getframerate(), fp.getnchannels(),fp.getnframes(),fp.getsampwidth
      (), fp.getcomptype(),fp.getcompname())
```

```python
#This is the play part
import pyaudio 
chunk = 1024
p = pyaudio.PyAudio()
stream = p.open(format = p.get_format_from_width(fp.getsampwidth()),
                channels = fp. getnchannels(), rate = fp.getframerate(),
                output = True)
data = fp.readframes(chunk)

while data !='':
    stream.write(data)
    data = fp.readframes(chunk)
    
stream.stop_stream()
stream.close()
p.terminate()
```
