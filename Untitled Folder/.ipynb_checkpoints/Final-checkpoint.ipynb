# Move to Top-hat transform

```python
import scipy.ndimage.morphology as morphology
import sys 
import numpy
from PIL import Image
```

```python
mg = morphology.morphological_gradient(img, (10,10,10))
imshow(mg)
```

White top-hat seems work for my intension, I tried to more.

```python
import sys
import numpy
from PIL import Image


im = numpy.array(img)
fft_mag = numpy.abs(numpy.fft.fftshift(numpy.fft.fft2(im)))

visual = numpy.log(fft_mag)
visual = (visual - visual.min()) / (visual.max() - visual.min())

result = Image.fromarray((visual * 255).astype(numpy.uint8))
result.save('/Users/intaehwang/Desktop/out.bmp')
```

```python
wtp = morphology.white_tophat(img, (15,15,15))
imshow(wtp);
```

```python
scr=wtp[2000:3000, 1000:2000]
subplot(121)
imshow(scr)
scr1=img[2000:3000, 1000:2000]
subplot(122)
imshow(scr1)
```
